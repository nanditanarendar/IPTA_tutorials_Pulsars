## setup the sampler and run it!

You'll want to run for a lot more samples than our short example.

If this is your first time using an MCMC consider this thought process:

>Most proposed "moves" will only update 1 or 2 parameters at at time. Since your model has $10$s of varying parameters, it wil take several "moves" before all of the parameters have changed to new values.  This means the result of two "moves" in succession are not statistically independent.  In this case it is reasonable to use the default thinning of `10`.  If you want to collect $50$k samples, then you'll need $500$k iterations.  It is common to set the `burn` to be about $1/10$ of the total iterations.

That means `Niter=500_000`, `thin=10`, `burn=50_000`.
